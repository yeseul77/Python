# account v1을 통해 game아이디#태그 -> PUUID 얻는 함수
# puuid를 통해 match_id를 얻어내는 함수 (20개중 한개만 호출)
# match_id를 통해 matches와 timeline 데이터를 얻어내는 함수

```python
api_key = 'RGAPI-3e834f86-cece-4e8a-9399-3b8110fcbe60'
```

```python
nickname = '건강한코어'
tag = 'KR1'
```

```python
def get_puuid(nickname,tag):
    url = f'https://asia.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{nickname}/{tag}?api_key={api_key}'
    res = requests.get(url).json()
    puuid = res['puuid']
    return puuid
```

```python
puuid = get_puuid(nickname,tag)
```

```python
puuid
```

```python
def get_match_id(puuid):
    url = f'https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count=20&api_key={api_key}'
    match_list = requests.get(url).json()
    return match_list
```

```python
match_list = get_match_id(puuid)
```

```python
match_list
```

```python
def get_matches_timelines(matchid):
    url1 = f'https://asia.api.riotgames.com/lol/match/v5/matches/{matchid}?api_key={api_key}'
    url2 = f'https://asia.api.riotgames.com/lol/match/v5/matches/{matchid}/timeline?api_key={api_key}'
    matches = requests.get(url1).json()
    timelines = requests.get(url2).json()
    return matches,timelines
```

```python
from tqdm import tqdm
import pandas as pd
```

```python
api_key = 'RGAPI-3e834f86-cece-4e8a-9399-3b8110fcbe60'
nickname = '건강한코어'
tag = 'KR1'
puuid = get_puuid(nickname,tag)
match_list = get_match_id(puuid)
data_list = []

for matchid in tqdm(match_list):
    matches,timelines = get_matches_timelines(matchid)
    data_list.append([matchid,matches,timelines])
```

```python
df = pd.DataFrame(data_list,columns = ['match_id','matches','timelines'])
```
